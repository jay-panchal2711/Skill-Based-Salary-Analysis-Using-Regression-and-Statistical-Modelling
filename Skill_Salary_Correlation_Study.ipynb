# Exploratory Data Analysis

#### 1. Mean & Median Salary by Skill

```python
skill_cols = ['python_yn', 'r_yn', 'sql_yn', 'excel_yn', 'spark_yn',
              'aws_yn', 'tableau_yn', 'hadoop_yn', 'sas_yn', 'java_yn']

skill_salary = []
for skill in skill_cols:
    mean_salary = df.loc[df[skill] == 1, 'avg_salary'].mean()
    median_salary = df.loc[df[skill] == 1, 'avg_salary'].median()
    skill_salary.append([skill.replace('_yn','').upper(), round(mean_salary,2), round(median_salary,2)])

skill_salary_df = pd.DataFrame(skill_salary, columns=['Skill', 'Mean Salary (USD)', 'Median Salary (USD)'])
skill_salary_df = skill_salary_df.sort_values(by="Mean Salary (USD)", ascending=False)

print("Mean & Median Salary by Skill:\n")
print(skill_salary_df)
```

#### 2. Mean & Median Salary by Job Role

```python
salary_by_role = (
    df.groupby('Job Title')['avg_salary']
    .agg(['mean', 'median'])
    .round(2)
    .sort_values('mean', ascending=False)
)

print("\n Mean & Median Salary by Job Title:\n")
print(salary_by_role.head(10))
```

#### 3. Mean & Median Salary by Location (State)

```python
salary_by_location = (
    df.groupby('job_state')['avg_salary']
    .agg(['mean', 'median'])
    .round(2)
    .sort_values('mean', ascending=False)
)

print("\n📍 Mean & Median Salary by Location:\n")
print(salary_by_location.head(10)) 
```

#### 4. Salary Distribution Across Experience Levels (using age)

#### 5. Identify Top-Earning Skill Combinations

```python
skill_cols = ['python_yn', 'r_yn', 'sql_yn', 'excel_yn', 'spark_yn', 
    'aws_yn', 'tableau_yn', 'hadoop_yn', 'sas_yn', 'java_yn']

def combine_skills(row):
    skills = [col.replace('_yn', '').upper() for col in skill_cols if row[col] == 1]
    return ' + '.join(skills) if skills else 'None'

df['skill_combo'] = df.apply(combine_skills, axis=1)
```

```python
skill_salary = (
    df.groupby('skill_combo')['avg_salary']
      .agg(['count', 'mean', 'median'])
      .sort_values(by='mean', ascending=False)
      .reset_index()
)
```

```python
top_combos = skill_salary[skill_salary['count'] >= 3].head(10)
```

```python
print("Top 10 Highest-Paying Skill Combinations:")
print(top_combos)
```

# Model building

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Define features and target
skill_cols = ['python_yn', 'r_yn', 'sql_yn', 'excel_yn', 'spark_yn',
              'aws_yn', 'tableau_yn', 'hadoop_yn', 'sas_yn', 'java_yn']
num_cols = ['age']             # Experience proxy
cat_cols = ['Industry']     

target = 'avg_salary'
```

```python
X = df[skill_cols + num_cols + cat_cols]
y = df[target]
```

```python
# Build the model preprocessing and Pipeline

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),      # Standardize age
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols),
    ('skills', 'passthrough', skill_cols)     # Keep skill flags as-is
])

model = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2025)
```

```python
# model fit

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"Model R²: {r2:.3f}")
```

# Visualization & Insights

#### 1. Bar Chart → Average Salary by Skill

```python
# List of skill columns
skills = ['python_yn', 'r_yn', 'sql_yn', 'excel_yn', 'spark_yn',
          'aws_yn', 'tableau_yn', 'hadoop_yn', 'sas_yn', 'java_yn']

# Compute average salary by skill
skill_salary = {skill: df.loc[df[skill] == 1, 'avg_salary'].mean() for skill in skills}
skill_salary = pd.Series(skill_salary).sort_values(ascending=False)

# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=skill_salary.values, y=skill_salary.index, palette="Blues_d")
plt.title("Average Salary by Skill (USD)", fontsize=14)
plt.xlabel("Average Salary (USD)")
plt.ylabel("Skill")
plt.show()
```

#### 2. Heatmap → Skill vs Experience vs Salary

```python
heatmap_data = df.pivot_table(values='avg_salary',
                                    index='age',
                                    columns='python_yn',
                                    aggfunc='mean')

plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, cmap="YlGnBu")
plt.title("Salary Heatmap: Experience (Age) vs Python Skill", fontsize=14)
plt.xlabel("Python Skill (0 = No, 1 = Yes)")
plt.ylabel("Company Age (proxy for experience)")
plt.show()
```

#### 3. Bubble Chart → Top Paying Skill Clusters

```python
skill_summary = pd.DataFrame({
    'Skill': skills,
    'Avg_Salary': [df[df[s]==1]['avg_salary'].mean() for s in skills],
    'Count': [df[s].sum() for s in skills]
})

plt.figure(figsize=(8,6))
plt.scatter(skill_summary['Count'], skill_summary['Avg_Salary'],
            s=skill_summary['Count']*10, alpha=0.6, c=skill_summary['Avg_Salary'], cmap='coolwarm')

for i in range(len(skill_summary)):
    plt.text(skill_summary['Count'][i]+1, skill_summary['Avg_Salary'][i],
             skill_summary['Skill'][i], fontsize=9)

plt.title("Top Paying Skill Clusters", fontsize=14)
plt.xlabel("Job Count (Skill Frequency)")
plt.ylabel("Average Salary (USD)")
plt.show()
```
